```python
lst = [1, 2, 3, 4]
```

```python
lst.append(5)
print(lst)
```

```python
# must be []
lst.extend([6, 7, 8])
print(lst)
```

```python
lst.insert(len(lst), 9)
print(lst)
```

```python
# remove the first that is 9 (not the 9th) 
lst.remove(9)
print(lst)
```

```python
# remove i (0 to len(lst-1))
lst.pop(7)
print(lst)
```

```python
lst1 = [i for i in lst]
lst1.clear()
print(lst1)
print(lst)
```

```python
# list.index(x, start,end)
# start and end are optional
lst.index(4, 1, 4)
```

```python
lst.count(7)
```

```python
lst2 = [1, 4, 2, 3]
lst2.sort()
print(lst2)
```

```python
lst3 = [1, 4, 2, 3]
lst3.reverse()
print(lst3)
```
