```python
for item in zip([1, 2, 3], ['sugar', 'spice', 'everything nice']):
    print(item)
```

```python
lst = list(zip(range(3), ['fee', 'fi', 'fo', 'fum']))
print(lst)

# Can not run with "strict=True" (use when iterables are equal in length):
# list(zip(range(3), ['fee', 'fi', 'fo', 'fum'], strict=True))
```

```python
for item in zip([1, 2, 3]):
    print(item)
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
print(list(zip(x, y)))
```

```python
x2, y2 = zip(*zip(x, y)) # (1, 4), (2, 5), (3, 6) -> (1, 2, 3), (4, 5, 6)
print(x2)
print(y2)
```
