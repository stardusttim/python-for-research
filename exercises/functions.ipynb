```python
def fib(n):
    p, q = 0, 1
    while p <= n:
        print(p, end=' ')
        p ,q = q, p+q
fib(2000)
```

```python
def ask_ok(prompt, tries=4, reminder='Please try agian, only yes or no'):
    while True:
        reply = input(prompt)
        if reply in ('y', 'ye', 'yes'):
            return True
        elif reply in ('n', 'no'):
            return False
        tries -= 1
        if tries <= 0:
            raise ValueError('Wrong')
        print(reminder)
print(ask_ok('Are you a man?'))
print(ask_ok('Did you have dinner?', 6, 'F'))
```

```python
def list_example(a, lst=None):
    if lst is None:
        lst = []
    lst.append(a)
    return lst
print(list_example(1))
print(list_example(2))
print(list_example(3))
print(list_example(2, [12]))
```

```python
def favorate(animal, color='black', city='NY'):
    print('Your favorite animal is:', animal, end='. ')
    print('Your favorite color is:', color, end='. ')
    print('Your favorite city is:', city+'.')
favorate('dog')
favorate(color='white', animal='cat')
favorate('elephant', 'blue', 'HK')
favorate('tiger', city='SG')
```

```python
#  * (tuple)
#  ** (dictionary)


def cheeseshop(kind, *arguments, **keywords):
    print(f'Do you have any {kind}?')
    print(f'Sorry, we are all out of {kind}.')
    for arg in arguments:
        print(arg)
    print("-" * 40)
    for position, name in keywords.items():
        print(f'{position}: {name}')
        
cheeseshop("Hamburger", 
           "It's very runny, sir.", "It's really very, VERY runny, sir.",
           shopkeeper="Michael Palin", client="John Cleese",sketch="Cheese Shop Sketch")
```

```python
my_args = ("It's very runny.", "It's really runny.")
my_kwargs = {'shopkeeper': 'Michael', 'client': 'John'}

cheeseshop("Limburger", *my_args, **my_kwargs)
```

```python
def standard_arg(arg):
    print(arg)
standard_arg(2)
```

```python
def pos_only_arg(arg1, arg2, /):
    print(arg1)
    print(arg2)
pos_only_arg(3, 5)
```

```python
def ked_only_arg(*,arg3, arg4):
    print(arg3)
    print(arg4)
ked_only_arg(arg3=4, arg4=6)
```

```python
def combined(pos_only, /, standard, *, ked_only):
    print(pos_only)
    print(standard)
    print(ked_only)
combined(1, 2, ked_only=3)
print('-'*5)
combined(1, standard=2, ked_only=3)
```
