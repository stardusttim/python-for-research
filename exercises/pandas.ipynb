```python
import pandas as pd

# Generate an employee data table containing missing values and unsorted data
raw_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance', 'IT'],
    'Salary': [5000, 8000, 7500, None, 9000, 8000],
    'Age': [25, 30, 28, 40, None, 35]
}

# Write to local CSV file
pd.DataFrame(raw_data).to_csv('data.csv', index=False)
```

```python
df = pd.read_csv('data.csv')
print(df)
```

```python
high_salary = df[df['Salary'] > 7500]
print(high_salary)
```

```python
it_dept = df[df['Department'] == 'IT']
print(it_dept)
```

```python
it_and_highsalary = df[(df['Salary'] > 7500) & (df['Department'] == 'IT')]
print(it_and_highsalary)
```

```python
df_sorted_age = df.sort_values(by='Age')
print(df_sorted_age)
```

```python
df_sorted_salary = df.sort_values(by='Salary', ascending=False)
print(df_sorted_salary)
```

```python
print(df.mean(numeric_only=True))
```

```python
print(df['Salary'].mean())
```

```python
# df.groupby('grouping column')['computed column'].function()

dept_salary = df.groupby('Department')['Salary'].mean()
print(dept_salary)
print('-' * 40)
print(dept_salary.sort_values(ascending=False))
```

```python
df_dropped = df.dropna()
print(df_dropped)
```

```python
df_dropped = df.dropna(subset='Salary')
print(df_dropped)
```

```python
df_filled = df.copy()
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].mean())
df_filled['Age'] = df_filled['Age'].fillna(0)
print(df_filled)
```

```python
df_filled.to_csv('cleaned_data.csv', index=False)
```
