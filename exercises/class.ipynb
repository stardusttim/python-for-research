```python
class Complex:
    def __init__(self, numerator, denominator):
        self.n = numerator
        self.d = denominator
y = Complex(1, 2)
print(y.n)
print(y.d)
n = y.n / y.d
print(n)
```

```python
class Dog:
    kind = 'canine'
    
    def __init__(self, name):
        self.name = name
        
d = Dog('Fido')
e = Dog('Buddy')
print(d.kind)
print(e.kind)
print(d.name)
print(e.name)
```

```python
class Dog:
    tricks = []
    
    def __init__(self, name):
        self.name = name
        
    def add_trick(self, trick):
        self.tricks.append(trick)
        
d = Dog('Fido')
e = Dog('Buddy')
d.add_trick('roll over')
e.add_trick('have fun')
print(d.tricks)
print(e.tricks)
```

```python
class Dog:   
    def __init__(self, name):
        self.name = name
        self.tricks = []
        
    def add_trick(self, trick):
        self.tricks.append(trick)
        
d = Dog('Fido')
e = Dog('Buddy')
d.add_trick('roll over')
e.add_trick('have fun')
print(d.tricks)
print(e.tricks)
```
