```python
data = np.array([
    [1, 5, 3],
    [4, 2, 6]
])

# 1. 不指定 axis：计算全局
print(np.sum(data))   # 21 (1+5+3+4+2+6)
print(np.mean(data))  # 3.5 (21 / 6)

# 2. axis=0（从上往下压，计算每列的结果）
print(np.sum(data, axis=0))  # [5, 7, 9] (1+4, 5+2, 3+6)
print(np.max(data, axis=0))  # [4, 5, 6] 

# 3. axis=1（从右往左压，计算每行的结果）
print(np.sum(data, axis=1))  # [9, 12] (1+5+3, 4+2+6)

# 4. argmax：返回最大值的“索引”位置
print(np.argmax(data))          # 5 (最大值是6，展平后的索引是5)
print(np.argmax(data, axis=0))   # [1, 0, 1] (第一列最大值是4在索引1，第二列是5在索引0...)
```

```python
data = np.array([
    [1, 5, 3],
    [4, 2, 6]
])

print(np.sum(data))
print(np.sum(data, axis=0))
print(np.sum(data, axis=1))
```

```python
print(np.mean(data))
print(np.mean(data, axis=0))
print(np.mean(data, axis=1, dtype=np.int16))
```

```python
print(np.max(data, axis=0))
print(np.min(data, axis=1))
```

```python
print(np.argmax(data, axis=0))  # Return index
print(np.argmax(data))
print(np.argmin(data, axis=1))
```
